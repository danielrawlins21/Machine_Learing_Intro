```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Crear el DataFrame con los datos proporcionados
df = pd.DataFrame({
    'horas': [1, 2, 4, 5, 5, 6, 7, 8, 10, 11, 11, 12, 12, 14, 15],
    'nota': [35, 47, 55, 56, 64, 76, 83, 84, 88, 89, 84, 73, 91, 93, 89]
})

print(df)
```

```python
# Crear la gráfica de dispersión
plt.scatter(df['horas'], df['nota'])
plt.title('Horas estudiadas vs. Nota del examen')
plt.xlabel('horas')
plt.ylabel('nota')

# Mostrar la gráfica
plt.show()
```

```python
# Model build
X = df['horas']
y = df['nota']

LR = LinearRegression()
LR.fit(X.array.reshape(-1,1),y)
```

```python
# value of x on the equation
print('Coefficients: \n', LR.coef_)

# Value where the Y axes is 0
print('Independent term: \n', LR.intercept_) 
```

```python
# Equation

y = LR.coef_*X + LR.intercept_
plt.plot(X, y, '-r')
plt.scatter(df.horas, df.nota)
plt.title('Horas estudiadas vs. Nota del examen')
plt.xlabel('horas')
plt.ylabel('nota')
plt.show()
```
