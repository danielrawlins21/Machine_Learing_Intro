```python
import numpy as np
import pandas as pd
from sklearn.cluster import kmeans_plusplus
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
x1 = np.random.standard_normal((100,2))* 0.6 + np.ones((100,2))
x2 = np.random.standard_normal((100,2))* 0.8 - np.ones((100,2))
x3 = np.random.standard_normal((100,2))* 0.4 - 1 * np.ones((100,2)) + 5
X = np.concatenate((x1,x2,x3), axis=0)
```

```python
plt.plot(X[:,0], X[:,1], 'k.' )
plt.show()
```

```python
n = 3
centers, labels = kmeans_plusplus(X=X,n_clusters=n)
```

```python
k_means = KMeans(n_clusters=n)
k_means.fit(X)
```

```python
centers
```

```python
labels
```

```python
cent = k_means.cluster_centers_
lab = k_means.labels_

print(cent,lab)
```

```python
plt.plot(X[lab==0,0],X[lab==0,1], 'g.', label='cluster 1')
plt.plot(X[lab==1,0],X[lab==1,1], 'r.', label='cluster 2')
plt.plot(X[lab==2,0],X[lab==2,1], 'b.', label='cluster 3')

plt.plot(centers[:,0], centers[:,1], 'mo', markersize=8, label='centroids')

plt.legend(loc='best')
plt.show()
```

```python
sum_squared_dist = []
k = range(1,10)

for k in k:
    km = KMeans(n_clusters=k, random_state=0)
    km = km.fit(X)
    sum_squared_dist.append(km.inertia_)

print(sum_squared_dist)
```

```python
plt.plot([1,2,3,4,5,6,7,8,9], sum_squared_dist)
plt.xlabel('Numero de clusters K')
plt.ylabel('Distancia media al centro del cluster en funcion de k')
```
