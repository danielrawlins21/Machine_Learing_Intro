```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de ejemplo basados en los resultados de Milagros Vergara
categories = ['D', 'I', 'S', 'C']
scores = [0, -1, 4, 5]
intensities = np.arange(-28, 29, 1)  # Rango de intensidades desde -28 a +28
```

```python
# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 8))

# Configurar la grilla de fondo
ax.set_facecolor('#eaf3fc')
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5, color='white')

# Aadir lneas verticales para las categoras
for i in range(len(categories)):
    ax.axvline(x=i + 1, color='lightgray', linewidth=1.5)

# Plotear los resultados
for i, score in enumerate(scores):
    ax.plot([i + 1] * len(intensities), intensities, marker='o', markersize=4, color='blue')
    ax.plot([i + 1], [score], marker='o', markersize=12, color='red')  # Destacar el puntaje real

# Etiquetas y ttulo
ax.set_xticks(np.arange(1, len(categories) + 1))
ax.set_xticklabels(categories)
ax.set_yticks(np.arange(-28, 29, 4))
ax.set_yticklabels(np.arange(-28, 29, 4))
ax.set_xlabel('Categoras DISC')
ax.set_ylabel('Intensidad')
ax.set_title('Grfica I - Resultados DISC de Milagros Vergara')

plt.show()
```
