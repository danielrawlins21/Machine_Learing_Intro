```python
import pandas as pd
import numpy as np
import glob

path = 'datasets'
files = glob.glob(path+'\*.csv')

#dataset = pd.read_csv('datasets\data.csv', sep=',')
```

```python
mapped_files = map(pd.read_csv, files)
dataset = pd.concat(mapped_files)
```

```python
data = dataset.copy()
```

```python
data.columns = data.columns.str.replace(' ', '_')
data.columns
```

```python
data.columns = data.columns.str.replace(r'[\^0-9]', '')
data.columns
```

```python
data.columns = data.columns.str.normalize('NFKD').str.encode('ascii',errors="ignore")
data.columns
```

```python
data.columns = data.columns.str.decode('utf-8')
data.columns
```
